```python
# lógica para calcular el area pedida

import numpy as np
import matplotlib
import matplotlib.pyplot as plt
import pandas as pd
from scipy.spatial import distance

#puntos del cuadrado
a = (0,0)
b = (1,0)
c = (1,1)
d = (0,1)

#distancias entre puntos 
dst1 = distance.euclidean(a, b)
dst2 = distance.euclidean(b, c)

Area = (dst1*dst2)/2
simulacion = 1000000

#lanzamiento de dardos
x = np.random.random(simulacion) #*100
y = np.random.random(simulacion) #*100

quarter = 1-(x**2)
half = 1-x

#dardos bajo y sobre las curvas
bajo_circulo = x<quarter
sobre_linea = y>half

#condicion logica, los dardos bajo el circulo y subre la linea 
z  = np.logical_and(bajo_circulo,sobre_linea)

suma = z.sum()
Area2 = Area*suma/simulacion


print('area sombreada:', Area2)

```
